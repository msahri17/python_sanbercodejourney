<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# Figure dan Axis kosong
fig, ax = plt.subplots()
plt.show()

# Plot Explicit
fig, ax = plt.subplots()
ax.plot(x,y,color='green')
ax.set_title('Plot Explicit', color='black')
plt.show()

# Plot implisit
plt.plot(x,y, color='Red')
plt.title('Plot Implicit', color='black')
plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
# Plot dengan style seaborn
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x,y, color='blue')
ax.plot(x,y2, color='green')
ax.plot(x,y3,color='red')
ax.set_title('Seaborn Style', color='green')
plt.style.use('seaborn')
plt.show()

# Save gambar
fig.savefig('Plot Soal 3.png', dpi=300)
```
