### Correlation

Covariance >> mengindikasikan arah hubungan 2 variabel (-tak hingga - tak hingga)

correlation coefficient >> mengindikasikan seberapa kuat hubungan 2 variabel (-1 hingga 1)

```python
import numpy as np
np.cov(siswa1, siswa2)
```

```python
np.corrcoef(siswa1, siswa2)
```

```python
r,p = scipy.stats.pearsonr(siswa1, siswa2)
print(r)
print(p)
```

```python
result = scipy.stats.linregress(siswa1, siswa2)
print(result.rvalue)
print(result.pvalue)
```

### Law of Large Number(LLN)

menyatakan bahwa semakin banyak data atau eksperimen maka semakin besar peluang mendekati nilai yang diharapkan

image.png

```python
expected_value=np.mean([0, 1])
expected_value
```

```python
import random
```

```python
random.choices([0,1], k=10)
```

```python
np.mean(random.choices([0,1], k=20))
```

```python
list_of_mean = []
list_of_k = np.arange(10, 5000, 10)
list_of_k
```

```python
for k in list_of_k:
    list_of_value = random.choices([0,1], k=k)
    mean = np.mean(list_of_value)
    list_of_mean.append(mean)
```

```python
list_of_mean
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18,12))
plt.plot(list_of_mean)
plt.hlines(0.5, 0, len(list_of_k))
plt.show()
```

```python
import numpy as np
from numpy.random import randn

penghitung = 0
N = 1000000

for i in randn(N):
    if i > -3 and i < 3:
        penghitung += 1
        
Hasil = penghitung/N
print(Hasil*100,'%')
```
