Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# membuat komponen figur dan axis serta menambahkan warna pada border(maaf kak, hanya sedikit variasi)
fig, ax = plt.subplots(facecolor='#F0FFFF',figsize=(12,8))

# Memberikan data pada axis 
ax.plot(x, cos, color='red', label='Cos X', ls='solid')
ax.plot(x, cos3, color='green', label='3 Cos X', ls='dashed')
ax.plot(x, sin, color='blue', label='sin(X-90)', ls='dashdot')
ax.plot(x, sin2, color='yellow', label='2 Sin X', ls='none', marker='x')

#memberikan axis dan title label
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos-Sin')

plt.legend(loc='center')

plt.show()

# Save Figure
fig.savefig('Coba.Png', dpi=500)
```
