```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots (1,3, figsize=(17,8), sharex=True, sharey=True)
ax[0].plot(x,tan, color='blue', label='tan X')
ax[1].plot(x,sin, color='red', label='sin X')
ax[2].plot(x,cos, color='green', label='cos X')

fig.legend(loc='upper center')
ax[0].set_title('Grafik tan X')
ax[1].set_title('Grafik sin X')
ax[2].set_title('Grafik cos X')


plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots (figsize=(16,10))
ax.plot(x,y)
fig.suptitle('Grafik cos X')

# untuk axis cos x^5
ax5 = fig.add_axes([0.055, 0.81, 0.22, 0.22])
ax5.plot(x,y,color='blue')
ax5.plot(x,y5, color='yellow')
ax5.set_title('Grafik cos X^5')

# untuk axis cos x^4
ax4 = fig.add_axes([0.825, 0.81, 0.22, 0.22])
ax4.plot(x,y, color='blue')
ax4.plot(x,y4, color='black')
ax4.set_title('Grafik cos X^4')

# untuk axis cos x^3
ax1 = fig.add_axes([0.83, 0.05, 0.22, 0.22])
ax1.plot(x,y,color='blue')
ax1.plot(x,y3, color='green')
ax1.set_title('Grafik cos X^3')

# untuk axis cos x^2
ax2 = fig.add_axes([0.055, 0.05, 0.22, 0.22])
ax2.plot(x,y,color='blue')
ax2.plot(x,y2, color='red')
ax2.set_title('Grafik cos X^2')

plt.show()
```
